```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Importing data
path = Path('../Machine_Learning_Project23/data/Spotify_Song_Attributes.csv')
df = pd.read_csv(path)

df.head()
```

```python
# Checking value counts for id column
df['id'].value_counts()
```

```python
# Dropping duplicate songs
df.drop_duplicates(subset=['id'],inplace=True)
```

```python
# Checking that duplicates were dropped 
df['id'].value_counts()
```

```python
# Dropping irrelevant columns
df.drop(columns=['artistName','type','id','uri','track_href','analysis_url','msPlayed','duration_ms'],inplace=True)
```

```python
# Checking for nulls 
df.info()
```

```python
# Dropping nulls
df.dropna(inplace=True)
```
